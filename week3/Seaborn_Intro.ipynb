```python
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/artur/ml_shared/week3/avocado.csv")
```

```python
df.head()
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df.head()
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=df, x="AveragePrice", y="Large Bags", palette="muted", hue="region")
```

```python
sns.set_theme(style="whitegrid")
_, ax = plt.subplots(figsize=(6,6))
sns.scatterplot(x="AveragePrice", y="Total Bags", data=df, ax=ax, hue="Total Volume")
```

```python
df.head()
```

```python
sns.set_theme(style="darkgrid")
_, ax = plt.subplots(figsize=(12,6))
sns.lineplot(x="Total Volume", y="Total Bags", data=df, ax=ax, hue="type")
```

```python
sns.displot(data=df, x="year", y="Total Bags", bins=50)
```

```python
df.head()
```

```python
sns.set_theme(style="ticks")
palette = sns.color_palette("rocket_r")
sns.relplot(data=df, x="Date", y="AveragePrice", kind="line", palette=palette)
```
