```python
import seaborn as sns
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
sns.set_theme(style="ticks")
sns.lmplot(data=df, x="AveragePrice", y="Large Bags", palette="muted", hue="region")
```
