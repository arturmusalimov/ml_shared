```python

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["Disease"] = df["Heart Disease"].apply(lambda x: 1 if x == "Presence" else 0)
```

```python
df["Gender"] = df["Sex"].apply(lambda x: "Male" if x == 1 else "Female")
```

```python
#model
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("Disease", axis=1),
        	                                        df["Disease"], test_size = 0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
df.describe()
```

```python
df.info()
```

```python
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediction))
```
