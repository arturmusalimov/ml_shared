```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("/home/artur/ml_shared/week4/cancer.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
print(df['diagnosis'].unique())
```

```python
df.drop("Unnamed: 32", axis=1, inplace=True)
```

```python
df.drop("id", axis=1, inplace=True)
```

```python
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap="viridis")
```

```python
df.head()
```

```python
df["diagnosis"] = pd.get_dummies(df["diagnosis"], drop_first=True)
```

```python
df.head()
```

```python
df.rename(columns={"diagnosis": "isHarmful"}, inplace=True)
```

```python
df.head()
```

```python
df.info()
```

```python
sns.set_style("whitegrid")
sns.countplot(x="isHarmful", data=df, palette="RdBu_r")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.drop("isHarmful", axis=1), df["isHarmful"], test_size=0.2, random_state=101)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
prediction = model.predict(X_test)
```

```python
print(prediction)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, prediction))
```
