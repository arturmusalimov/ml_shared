```python
import numpy as np
import pandas as pd 
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
train_df = pd.read_csv("/home/artur/ml_shared/week6/svm_train.csv")
```

```python
train_df["Activity"].value_counts()
```

```python
class_names = {"LAYING":0, 
               "STANDING":1,
               "SITTING":2,
               "WALKING":3,
               "WALKING_UPSTAIRS":4,
               "WALKING_DOWNSTAIRS":5
               }
```

```python
X = train_df.drop("Activity", axis=1)
y = train_df["Activity"]
```

```python
y = train_df["Activity"].map(class_names)
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
```

```python
svm_model = SVC()
svm_model.fit(X_train, y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, y_pred))
print("-------------")
print(classification_report(y_test, y_pred))
```

# GRID SEARCH

```python
param_grid = {"C": [0.1, 1, 10, 100, 1000],
              "gamma": [1, 0.1, 0.01, 0.001, 0.0001] }
```

```python
grid_model = GridSearchCV(SVC(), param_grid, verbose=3)
grid_model.fit(X_train, y_train)
```

```python
grid_model.best_score_
```

```python
grid_model.best_params_
```

```python
grid_model.best_estimator_
```

```python
grid_preds = grid_model.predict(X_test)
```

```python
print(confusion_matrix(y_test, grid_preds))
print("-------------")
print(classification_report(y_test, grid_preds))
```
